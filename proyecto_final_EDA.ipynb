GRAFICAMOS MEDIDAS CENTRALES Y DE DISPERSIÓN DE LA VARIABLE DE LA "ENERGIA"

Energía

Es una medida de 0,0 a 1,0 y representa una medida perceptual de intensidad y actividad. Normalmente, las pistas energéticas se sienten rápidas, ruidosas y ruidosas. Por ejemplo, Death Metal tiene alta energía, mientras que un preludio de Bach obtiene bajas en la escala. Las características perceptivas que contribuyen a este atributo incluyen rango dinámico, ruido percibido, timbre, tasa de inicio y entropía general.

```python
sns.histplot(
    data=df_spotify_youtube,
    x='energy'
)
plt.axvline(
    x=df_spotify_youtube.energy.mean(),
    color='red',
    linestyle='dashed',
    linewidth=2,
    label="media"
)
plt.axvline(
    x=df_spotify_youtube.energy.median(),
    color='green',
    linestyle='dashed',
    linewidth=2,
    label="mediana"
)

plt.axvline(
    x=df_spotify_youtube.energy.quantile(0.25),
    color='yellow',
    linestyle='dashed',
    linewidth=2,
    label="Q1"
)
plt.axvline(
    x=df_spotify_youtube.energy.quantile(0.75),
    color='yellow',
    linestyle='dashed',
    linewidth=2,
    label="Q3"
)
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df_spotify_youtube.energy)
plt.title("Energia que genera la música")

tmp_mean = df_spotify_youtube.energy.mean()
plt.axvline(tmp_mean, linestyle="dashed", color="red", label=f"Media: {tmp_mean}")
plt.xlabel('')
plt.legend()
plt.show()
```
